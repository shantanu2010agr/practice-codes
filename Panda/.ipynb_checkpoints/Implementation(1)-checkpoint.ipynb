<ol start="8">
    <li> Filtering rows by column values</li></ol>

```python
movies=pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
booleans=[]
for length in movies.duration:
    if length >=200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
checkl= pd.Series(booleans)
```

```python
checkl.head()
```

```python
movies['check']=checkl
```

```python
movies.head()
```

```python
movies.drop('check',axis=1,inplace=True)
```

```python
# Final step of first method
movies[checkl]
# returns the df with all the movies having time duration 
# greater or equal to 200 mins.
```

```python
# Easier method as follows:
movies[movies.duration>=200]
# Inside bracket it is creating a series of True/False 
# & is passed to the df to get the desired result.
```

```python
# Bonus tip:
# Use of loc
movies.loc[movies.duration>=200,'genre']
# First part inside the bracket will help in selecting the rows
# Similarly, second side is for columns.
```
