```python
train=pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
feature_cols=['Pclass','Parch']
```

```python
X=train.loc[:,feature_cols]
```

```python
y= train.Survived
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
```

```python
# Its about creating classic ML model using scikit-leearn
logreg=LogisticRegression()
logreg.fit(X,y)
```

```python
test=pd.read_csv('http://bit.ly/kaggletest')
```

```python
test.head()
```

```python
Xp=test.loc[:,feature_cols]
```

```python
newS=logreg.predict(Xp)
```

```python
newS
```

```python
newS.shape
```

```python
# This will output a csv df file
# in the current working directory in my desktop
output=pd.DataFrame({'PassengerID':test.PassengerId,
                     'Survived':newS}).set_index('PassengerID').to_csv('out.csv')
```

```python
train=pd.read_csv('http://bit.ly/kaggletrain')
```

```python
# Mapping the data
train['Sex_male']=train.Sex.map({'female':0,'male':1})
```

```python
train.head()
```

```python
pd.get_dummies(train.Sex)
```

```python
# Forming the same df as same as "map" method.
pd.get_dummies(train.Sex,prefix='Sex').iloc[:,1:]
# We neeed n-1 dummy variable to represent them all the categories.
```

```python
# We can go for the index with more than one categories
train.Embarked.value_counts()
```

```python
embarked_dummies=pd.get_dummies(train.Embarked, prefix='Embarked').iloc[:,1:]
train=pd.concat([train, embarked_dummies],axis=1)
```

```python
train.head()
```

```python
#  Now not even requiring iloc, prefix , concat,
# to get the resulting df
pd.get_dummies(train, 
                columns=['Pclass','Survived'], drop_first=True)
```
