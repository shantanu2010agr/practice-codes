# Working in histogram:

```python
# use plt.hist() instead of plt.plot()

import random

n=14
bins=[]
popu=[]
nopop=[  ]

for i in range(n):
    bins.append(i*10)
for i in range(5*n):    
    popu.append(random.randrange(1,10*n))
    nopop.append(random.randrange(1,10*n))

print(bins)
print(popu)
# histtypes:'bar', 'barstacked', 'step', 'stepfilled'    
plt.hist(popu , bins ,histtype='bar',rwidth=0.6,label='One')
plt.hist(nopop , bins ,histtype='barstacked',rwidth=0.9,label='Five',color='g')

plt.title("Try Histogram")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True,color='k')

plt.legend()
plt.show()
```

# Working on Stack-Plot:

It is just a different numbers of plots being stacked up.<br>
So, basically, we plot up for all the categories(using same plt.plot()), then stacking up all the plots using plt.satckplot() (which stacks up all the plots, being taken).

```python
n=5
days=[]

sleeping=[]
eating=[]
working=[]
playing=[]

for i in range(n):
    days.append(i+1)

for i in range(n):
    d=random.randrange(1,23)
    a=random.randrange(1,23)
    b=random.randrange(1,23)
    c=random.randrange(1,23)
    while sum([a,b,c,d])!=24:
        d=random.randrange(1,23)
        a=random.randrange(1,23)
        b=random.randrange(1,23)
        c=random.randrange(1,23)
    sleeping.append(a)
    eating.append(b)
    working.append(c)
    playing.append(d)
    
plt.plot([],[],color='r',label='Sleeping', linewidth=5)
plt.plot([],[],color='g',label='Eating', linewidth=4)
plt.plot([],[],color='b',label='Working', linewidth=8)
plt.plot([],[],color='c',label='Playing', linewidth=2)

plt.stackplot(days,sleeping,eating,working,playing,colors=['r','g','b','c'])
    
plt.title("Try Stack-Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True,color='k')    

plt.legend()
plt.show()
```

# Working on pie-charts:

```python
sleeping=[]
eating=[]
working=[]
playing=[]

for i in range(n):
    d=random.randrange(1,23)
    a=random.randrange(1,23)
    b=random.randrange(1,23)
    c=random.randrange(1,23)
    while sum([a,b,c,d])!=24:
        d=random.randrange(1,23)
        a=random.randrange(1,23)
        b=random.randrange(1,23)
        c=random.randrange(1,23)
    sleeping.append(a)
    eating.append(b)
    working.append(c)
    playing.append(d)

activities=[sleeping,eating,working,playing]
act_name=['sleeping','eating','working','playong']
colors=['c','m','r','b']

print(sleeping)
print(eating)
print(working)
print(playing)

for i in range(n):
    slices=[]
    
    for j in activities:
        slices.append(j[i])
    
    plt.pie(slices,colors=colors,labels=act_name,
            shadow=True,explode=(0,0.2,0,0.1),
            startangle=45,
           autopct='%1.4f%%')
    
    plt.title("Pie-Chart")
    plt.legend()
    plt.show()
```

# Dealing with Multiple-Plots:

Using Numpy as well<br>
We'll use plt.subplot(abc)<br>
<ol><li>a=number of graphs</li>
    <li>b=horizontally avaialble graphs</li>
    <li>c=graph number on vertical axis<br>(or correspondingly)</li>
</ol>


```python
import numpy as np
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.title("Mutliple graphs")

plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Multiple functioning at once")

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
# plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("simple cos")

plt.show()
```
